```python
# Cell 3: Sắp xếp theo Recall
results_sorted_recall = results.sort_values(by="Recall", ascending=False).reset_index(drop=True)
display(results_sorted_recall)
```

```python
# Cell 4: Sắp xếp theo F1-score
results_sorted_f1 = results.sort_values(by="F1-score", ascending=False).reset_index(drop=True)
display(results_sorted_f1)
```

```python
# Cell 5: Sắp xếp theo ROC-AUC
results_sorted_auc = results.sort_values(by="ROC-AUC", ascending=False).reset_index(drop=True)
display(results_sorted_auc)
```

```python
# Cell 6: Sắp xếp theo Accuracy
results_sorted_acc = results.sort_values(by="Accuracy", ascending=False).reset_index(drop=True)
display(results_sorted_acc)
```

```python
# Cell 7: So sánh Accuracy
plt.figure(figsize=(10, 5))
plt.bar(results["Model"], results["Accuracy"])
plt.title("So sánh Accuracy giữa các mô hình")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Cell 8: So sánh Precision
plt.figure(figsize=(10, 5))
plt.bar(results["Model"], results["Precision"])
plt.title("So sánh Precision giữa các mô hình")
plt.xlabel("Model")
plt.ylabel("Precision")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Cell 9: So sánh Recall
plt.figure(figsize=(10, 5))
plt.bar(results["Model"], results["Recall"])
plt.title("So sánh Recall giữa các mô hình")
plt.xlabel("Model")
plt.ylabel("Recall")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Cell 10: So sánh F1-score
plt.figure(figsize=(10, 5))
plt.bar(results["Model"], results["F1-score"])
plt.title("So sánh F1-score giữa các mô hình")
plt.xlabel("Model")
plt.ylabel("F1-score")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Cell 11: So sánh ROC-AUC
plt.figure(figsize=(10, 5))
plt.bar(results["Model"], results["ROC-AUC"])
plt.title("So sánh ROC-AUC giữa các mô hình")
plt.xlabel("Model")
plt.ylabel("ROC-AUC")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Cell 12: Model tốt nhất theo từng metric
best_accuracy_model = results.loc[results["Accuracy"].idxmax(), ["Model", "Accuracy"]]
best_precision_model = results.loc[results["Precision"].idxmax(), ["Model", "Precision"]]
best_recall_model = results.loc[results["Recall"].idxmax(), ["Model", "Recall"]]
best_f1_model = results.loc[results["F1-score"].idxmax(), ["Model", "F1-score"]]
best_auc_model = results.loc[results["ROC-AUC"].idxmax(), ["Model", "ROC-AUC"]]

print("Best Accuracy:")
print(best_accuracy_model)
print()

print("Best Precision:")
print(best_precision_model)
print()

print("Best Recall:")
print(best_recall_model)
print()

print("Best F1-score:")
print(best_f1_model)
print()

print("Best ROC-AUC:")
print(best_auc_model)
```

```python
# Cell 13: Làm tròn bảng kết quả
results_rounded = results.copy()

metric_cols = ["Accuracy", "Precision", "Recall", "F1-score", "ROC-AUC"]
results_rounded[metric_cols] = results_rounded[metric_cols].round(4)

display(results_rounded)
```

```python
# Cell 14: Lưu bảng kết quả
results_rounded.to_csv("../outputs/model_comparison_results.csv", index=False)
print("Đã lưu bảng so sánh tại: ../outputs/model_comparison_results.csv")
```

```python
# Cell 15: Tính điểm trung bình các metric
results_rank = results.copy()
results_rank["Average_Score"] = results_rank[["Accuracy", "Precision", "Recall", "F1-score", "ROC-AUC"]].mean(axis=1)

results_rank = results_rank.sort_values(by="Average_Score", ascending=False).reset_index(drop=True)
display(results_rank)
```

```python
# Cell 16: Lưu bảng xếp hạng tổng hợp
results_rank.to_csv("../outputs/model_ranking_average_score.csv", index=False)
print("Đã lưu bảng xếp hạng tại: ../outputs/model_ranking_average_score.csv")
```

```python
# Cell 17: Tính điểm ưu tiên Recall
results_weighted = results.copy()

results_weighted["Weighted_Score"] = (
    results_weighted["Recall"] * 0.35 +
    results_weighted["F1-score"] * 0.25 +
    results_weighted["ROC-AUC"] * 0.20 +
    results_weighted["Precision"] * 0.10 +
    results_weighted["Accuracy"] * 0.10
)

results_weighted = results_weighted.sort_values(by="Weighted_Score", ascending=False).reset_index(drop=True)
display(results_weighted)
```

```python
# Cell 18: Lưu bảng xếp hạng ưu tiên Recall
results_weighted.to_csv("../outputs/model_ranking_weighted_score.csv", index=False)
print("Đã lưu bảng xếp hạng ưu tiên Recall tại: ../outputs/model_ranking_weighted_score.csv")
```

```python
# Cell 19: Biểu đồ Weighted Score
plt.figure(figsize=(10, 5))
plt.bar(results_weighted["Model"], results_weighted["Weighted_Score"])
plt.title("So sánh Weighted Score giữa các mô hình")
plt.xlabel("Model")
plt.ylabel("Weighted Score")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
