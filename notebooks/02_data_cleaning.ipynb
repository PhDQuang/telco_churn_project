```python
# Cell 1: Import thư viện
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
# Cell 2: Đọc dữ liệu gốc
file_path = "../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv"
df = pd.read_csv(file_path)

print("Kích thước ban đầu:", df.shape)
display(df.head())
```

```python
# Cell 3: Kiểm tra nhanh
print("Số dòng trùng hoàn toàn:", df.duplicated().sum())
print("Số customerID bị trùng:", df["customerID"].duplicated().sum())
print("Số dòng TotalCharges rỗng:", (df["TotalCharges"].astype(str).str.strip() == "").sum())
```

```python
# Cell 4: Thay chuỗi rỗng bằng NaN
df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)

print("Số NaN trong TotalCharges sau khi thay:", df["TotalCharges"].isnull().sum())
```

```python
# Cell 5: Chuyển TotalCharges sang numeric
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

print(df["TotalCharges"].dtype)
print("Số NaN trong TotalCharges sau khi convert:", df["TotalCharges"].isnull().sum())
```

```python
# Cell 6: Xem các dòng bị thiếu TotalCharges
missing_total = df[df["TotalCharges"].isnull()]
display(missing_total)
```

```python
# Cell 7: Xóa các dòng thiếu TotalCharges
df = df.dropna(subset=["TotalCharges"])

print("Kích thước sau khi xóa missing TotalCharges:", df.shape)
print("Số NaN còn lại trong TotalCharges:", df["TotalCharges"].isnull().sum())
```

```python
# Cell 8: Bỏ customerID
df = df.drop(columns=["customerID"])

print("Kích thước sau khi bỏ customerID:", df.shape)
print(df.columns.tolist())
```

```python
# Cell 9: Kiểm tra lại kiểu dữ liệu
print(df.info())
```

```python
# Cell 10: Kiểm tra missing sau cleaning
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    "So_luong_thieu": missing,
    "Ti_le_%": missing_percent
}).sort_values(by="So_luong_thieu", ascending=False)

display(missing_df)
```

```python
# Cell 11: Kiểm tra lại Churn
print(df["Churn"].value_counts())
print("\nTỉ lệ phần trăm:")
print(df["Churn"].value_counts(normalize=True) * 100)
```

```python
# Cell 12: Lưu file sạch
output_path = "../data/processed/telco_churn_clean.csv"
df.to_csv(output_path, index=False)

print(f"Đã lưu file sạch tại: {output_path}")
```
