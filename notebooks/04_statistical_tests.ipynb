```python
# Cell 1: Import thư viện
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, ttest_ind, mannwhitneyu
```

```python
# Cell 2: Đọc dữ liệu sạch
df = pd.read_csv("../data/processed/telco_churn_clean.csv")
print(df.shape)
display(df.head())
```

```python
# Cell 3: Hàm kiểm định chi-square
def chi_square_test(df, col, target="Churn"):
    contingency_table = pd.crosstab(df[col], df[target])
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    print(f"\n=== Chi-square test: {col} vs {target} ===")
    print("Bảng tần số:")
    display(contingency_table)

    print(f"Chi2 statistic = {chi2:.4f}")
    print(f"p-value = {p:.6f}")
    print(f"Degrees of freedom = {dof}")

    if p < 0.05:
        print("Kết luận: Bác bỏ H0 -> Có mối liên hệ có ý nghĩa thống kê.")
    else:
        print("Kết luận: Chưa đủ bằng chứng để bác bỏ H0.")
```

```python
# Cell 4: Kiểm định Contract và Churn
chi_square_test(df, "Contract")
```

```python
# Cell 5: Kiểm định TechSupport và Churn
chi_square_test(df, "TechSupport")
```

```python
# Cell 5: Kiểm định TechSupport và Churn
chi_square_test(df, "TechSupport")
```

```python
# Cell 7: Tách nhóm churn và non-churn
yes_group = df[df["Churn"] == "Yes"]
no_group = df[df["Churn"] == "No"]

print("Số lượng churn = Yes:", len(yes_group))
print("Số lượng churn = No :", len(no_group))
```

```python
# Cell 8: Kiểm định tenure giữa hai nhóm
stat, p = mannwhitneyu(yes_group["tenure"], no_group["tenure"], alternative="two-sided")

print("=== Mann-Whitney U test: tenure theo Churn ===")
print(f"Statistic = {stat:.4f}")
print(f"p-value = {p:.6f}")

if p < 0.05:
    print("Kết luận: tenure khác biệt có ý nghĩa thống kê giữa hai nhóm.")
else:
    print("Kết luận: chưa thấy khác biệt có ý nghĩa thống kê.")
```

```python
# Cell 9: Kiểm định MonthlyCharges giữa hai nhóm
stat, p = mannwhitneyu(yes_group["MonthlyCharges"], no_group["MonthlyCharges"], alternative="two-sided")

print("=== Mann-Whitney U test: MonthlyCharges theo Churn ===")
print(f"Statistic = {stat:.4f}")
print(f"p-value = {p:.6f}")

if p < 0.05:
    print("Kết luận: MonthlyCharges khác biệt có ý nghĩa thống kê giữa hai nhóm.")
else:
    print("Kết luận: chưa thấy khác biệt có ý nghĩa thống kê.")
```
